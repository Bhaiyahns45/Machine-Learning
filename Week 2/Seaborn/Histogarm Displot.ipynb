```python
from scipy.stats import norm
```

```python
sns.distplot(df.total_bill,axlabel="Total Bill")
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]

sns.distplot(df.total_bill,label="Total Bill", bins=bins, rug=True)

plt.xticks(bins)
plt.show()
```
