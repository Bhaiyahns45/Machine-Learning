```python
plt.figure(figsize=(16,9))
sns.heatmap(df, vmin=0, vmax=21, cmap="coolwarm", annot=True)
```

```python
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']], dtype = str)
 
sns.heatmap(arr, annot = annot_arr, fmt="s") # s -string, d - decimal
```

```python
plt.figure(figsize=(16,9))
annot_kws={'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.9, 
#            'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}

sns.heatmap(df, vmin=0, vmax=21, cmap="coolwarm", annot=True, annot_kws= annot_kws)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df, vmin=0, vmax=21, cmap="coolwarm", annot=True, linewidths=2)
```

```python
plt.figure(figsize=(16,9))
 
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
 
ax=sns.heatmap(df, vmin=0, vmax=21, cmap="coolwarm", annot=True, linewidths=2, yticklabels = country_code)

ax.set(title="Heatmap",
      xlabel="Years",
      ylabel="Country Name")

```
