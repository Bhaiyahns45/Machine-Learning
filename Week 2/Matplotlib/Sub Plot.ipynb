```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
languages=['Python', 'Java', 'Javascript', 'C', 'C++']
class1_stu=np.random.randint(10,20,(5))
plt.pie(class1_stu, labels=languages)

plt.subplot(2,2,2)
languages=['Python', 'Java', 'Javascript', 'C', 'C++']
class1_stu=np.random.randint(10,20,(5))
plt.bar(languages, class1_stu, edgecolor='r')
plt.xlabel("Language", fontsize=20)
plt.ylabel("No of Students", fontsize=20)

plt.subplot(2,2,3)
df=pd.read_csv('D:\Data_sets\Mall_Customers.csv')
df.rename(columns ={'Spending Score (1-100)':'Spending', 'Annual Income (k$)':'Income'}, inplace=True)

x= df.Age
y=df.Spending

plt.scatter(x,y)
plt.xlabel("Age", fontsize=20)
plt.ylabel("Spending", fontsize=20)

plt.subplot(2,2,4)
ml_students_age= np.random.randint(18,45,(100))
plt.hist(ml_students_age, bins=[15,20,25,30,35,40,45], rwidth=0.8, label="ML Students")
plt.legend()
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No of Students")


plt.show()
```
