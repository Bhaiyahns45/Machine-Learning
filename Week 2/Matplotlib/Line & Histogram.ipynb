### Histograms

```python
import random
```

```python
ml_students_age= np.random.randint(18,45,(100))
py_students_age= np.random.randint(15,44,(100))
```

```python
ml_students_age
```

```python
py_students_age
```

```python
plt.hist(ml_students_age)
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No of Students")
```

```python
plt.hist(ml_students_age, bins=[15,20,25,30,35,40,45], rwidth=0.8, label="ML Students")
plt.legend()
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No of Students")
```

```python
plt.hist(ml_students_age, bins=[15,20,25,30,35,40,45], rwidth=0.8, orientation="horizontal")
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No of Students")
```

```python
plt.figure(figsize=(16,9))
plt.hist([ml_students_age, py_students_age], bins=[15,20,25,30,35,40,45], rwidth=0.8,color=["r","y"], label=["ML Students", "PY Students"])
plt.legend()
plt.title("ML Students Age Histogram")
plt.xlabel("Age")
plt.ylabel("No of Students")
plt.show()
```
