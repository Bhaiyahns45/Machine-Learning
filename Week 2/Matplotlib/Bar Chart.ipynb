```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
languages=['Python', 'Java', 'Javascript', 'C', 'C++']

class1_stu=np.random.randint(10,20,(5))
class2_stu=np.random.randint(10,20,(5))
class3_stu=np.random.randint(10,20,(5))
```

```python
plt.bar(languages, class1_stu, edgecolor='r')
plt.xlabel("Language", fontsize=20)
plt.ylabel("No of Students", fontsize=20)
```

```python
languange_index=np.arange(len(languages))
width=0.2
plt.figure(figsize=(16,9))
plt.bar(languange_index, class1_stu, width, edgecolor='r', color="r")

plt.bar(languange_index+width, class2_stu, width, edgecolor='r', color="b")

plt.bar(languange_index+width+width, class3_stu, width, edgecolor='r', color="g")

plt.xticks(languange_index+width, languages)
plt.xlabel("Language", fontsize=20)
plt.ylabel("No of Students", fontsize=20)
```
