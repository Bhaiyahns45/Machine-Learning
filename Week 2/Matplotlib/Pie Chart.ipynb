```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
languages=['Python', 'Java', 'Javascript', 'C', 'C++']

class1_stu=np.random.randint(10,20,(5))
```

```python
class1_stu
```

```python
plt.pie(class1_stu, labels=languages)
plt.show()
```

```python
plt.pie(class1_stu, labels=languages, explode=[0.1,0,0,0,0])
plt.show()
```

```python
plt.pie(class1_stu, labels=languages, autopct="%0.2f%%")
plt.show()
```

```python
plt.pie(class1_stu, labels=languages, autopct="%0.2f%%", shadow=True)
plt.show()
```

```python
plt.pie(class1_stu, labels=languages, autopct="%0.2f%%", shadow=True, radius=1.5)
plt.show()
```

```python
plt.pie(class1_stu, labels=languages, autopct="%0.2f%%", shadow=True, radius=1.5, textprops={"fontsize":15})
plt.show()
```

```python
plt.pie(class1_stu, labels=languages, autopct="%0.2f%%", shadow=True, radius=1.5, textprops={"fontsize":15})
plt.legend(loc='upper right', bbox_to_anchor=(2, 1),fontsize=15)
plt.show()
```
