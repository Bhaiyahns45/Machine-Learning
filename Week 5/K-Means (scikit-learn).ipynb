```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(centers=3, n_samples=200, n_features=2, shuffle=True, random_state=40)
```

```python
# X[:,1]
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(1,5):
    kmeans=KMeans(n_clusters=i,random_state=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,5), wcss)
plt.xlabel('K')
plt.ylabel('WCSS')
```

## Example 2


```python
X = np.array([[5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,30],
    [60,70],
    [70,80],
    [74,65],
    [80,55],
    [85,91],])
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X[:,0],X[:,1])
```

```python
wcss=[]

for i in range(1,5):
    kmeans=KMeans(n_clusters=i,random_state=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,5), wcss)
plt.xlabel('K')
plt.ylabel('WCSS')
```

```python
k=KMeans(n_clusters=2)
```

```python
cl=k.fit_predict(X)
```

```python
cl
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(X[:,0],X[:,1], hue=cl)
```
