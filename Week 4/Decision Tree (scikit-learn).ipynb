```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from numpy import random
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
from sklearn.datasets import load_iris

iris_data=load_iris()

df=pd.DataFrame(iris_data.data)
df.columns=iris_data.feature_names
df['type']=iris_data.target
df.tail(10)
```

```python
df.shape
```

```python
X=df.drop('type', axis=1)
```

```python
y=df.type
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
dc=DecisionTreeClassifier(criterion='entropy', max_depth=4)
```

```python
dc.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(20,10))
plot_tree(dc)
plt.show()
```

```python
print(tree.export_text(dc))
```

```python
dc.score(X_test,y_test)
```

### Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
dc1=DecisionTreeClassifier()
```

```python
parameter= {'criterion':['gini','entropy'],
        'max_depth':[2,3,4,5,6],
        'min_samples_split':[1,2,3,4],
        'min_samples_leaf':[1,2],
          }
```

```python
dc_grid=GridSearchCV(dc1,parameter,cv=5)
```

```python
dc_grid.fit(X_train,y_train)
```

```python
dc_grid.best_params_
```

```python
dc_grid.score(X_test,y_test)
```
